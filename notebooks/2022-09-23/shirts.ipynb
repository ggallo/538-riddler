# Selling Shirts

A store lists N shirts to buy, labeled 1 to $N$. The buyers somehow aren't told anymore details about the product, and need to keep buying, returning, then buying until they get their desired shirt. They do this at random (and don't talk to each other apparently), with elimination i.e. they won't buy the same shirt again.

The store gets has an 80% average return rate. What is $N$?

We can simulate. WLOG assume everyone wants the shirt labeled 0. We'll try samples of 100000 for each value of $N$ up to 20 inclusive. This won't be efficient.

```python
import seaborn as sns
from random import shuffle
average_d ={}
for i in range(2, 13):
    returns, total = 0, 0
    for j in range(100000):
        b = list(range(i))
        shuffle(b)
        returns += b.index(0)
        total += b.index(0) + 1
    average_d[i] = 1. * returns / total
```

```python
ax = sns.barplot(x=list(average_d.keys()), y=list(average_d.values()))
ax.grid(axis='y')
for i in ax.containers:
    ax.bar_label(i,fmt='%.2f')
```

So there's 9 shirts, but just a few minutes of pen and paper would have solved this problem for any percentage. For $N$ shirts, the average percentage of orders that are returns will be the (n-1)-th triangular number over the nth, which simplifies to:

$$
f(n) = \frac{n-1}{n+1}
$$
